```python
# softmax
def softmax(pre_layer): # 'pre_layer' specifies sum of weights and biases before softmax is applied
    a = 1 / np.sum(np.exp(pre_layer)) # will need to compute for some batch size all at once
    b = np.exp(pre_layer)
    return a * b
```
